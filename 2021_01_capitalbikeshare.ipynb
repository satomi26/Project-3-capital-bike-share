first we reduce the height and width of the map, and then we'll change the zoom level. we can resize our map by using the branca library in Python. It is spin off from folium that hosts the non-map specific features. We can use its Figure class for resizing our maps and pass the desired width and height in pixels.
https://www.analyticsvidhya.com/blog/2020/06/guide-geospatial-analysis-folium-python/

```python
from branca.element import Figure
fig=Figure(width=550,height=350)
```

Next, let’s move to our second problem and change the zoom level. By default, you can zoom in and out as much as you want. Also, the starting zoom level is fixed to 10 (the map is plotted with a 10x zoom). Now, you might be thinking – how can this be a problem?

Well, imagine you want to plot a map of a small area. Here, you need to mention the appropriate starting zoom level so that the map focuses only on that region. Also, if you don’t want them to zoom in or out much and lose focus on the map, then you can restrict it. For doing this, Folium gives us three parameters – zoom_start, min_zoom, and max_zoom.

So, let us again plot the map of Delhi but this time of fixed size and with a restricted zoom level:

```python
m1=folium.Map(width=550,height=350,location=[38.947156, -77.065115],zoom_start=11,min_zoom=8,max_zoom=14)
fig.add_child(m1)
m1
```

```python

# Creating Basemap
fig3=Figure(width=550,height=350)
m3=folium.Map(location=[38.947156, -77.065115],tiles='cartodbpositron',zoom_start=10)
fig3.add_child(m3)

#Adding markers to the map
folium.Marker(location=[38.947156, -77.065115],popup="<strong>Start point</strong>",tooltip='Click here to see Popup', icon = folium.Icon(icon = "cloud")).add_to(m3)
folium.Marker(location=[38.949662, -77.027333],popup='<strong>End point</strong>',tooltip='<strong>Click here to see Popup</strong>', icon = folium.Icon(icon = "envelope", color = "red") ).add_to(m3)

m3
```

```python
bike_map = folium.Map(location=[38.907192,-77.036873], zoom_start=5) # location = washington DC
fg = folium.FeatureGroup()
for i, row in df_slice.iterrows():
  fg.add_child(folium.Marker(
               location = [row["start_lat"], row["start_lng"]],

               popup = row["start_station_name"],
               icon = folium.Icon(color = "blue")
  ))
bike_map.add_child(fg)
bike_map
```
